```python
def preprocessing(df):
    missing = 'Missing'
    data = df.copy()
    data.fillna({'Alley': 'No alley', 'Fence': 'No fence', 'FireplaceQu': 'No fireplace', 'MasVnrType': missing, 'Electrical': missing}, inplace=True)
    data[['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']] = data[['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']].fillna(value='No garage')
    data[['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']] = data[['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']].fillna(value='No basement')
    data[['GarageYrBlt', 'LotFrontage', 'MasVnrArea']] = data[['GarageYrBlt', 'LotFrontage', 'MasVnrArea']].fillna(value=-999)
    data.drop('PoolQC', axis=1, inplace=True)
    data['MiscFeature'] = data['MiscFeature'].replace({None: missing, 'Shed': 'Shed', 'Gar2': missing, 'Othr': missing, 'TenC': missing})

    categorical_columns = list(data.select_dtypes(exclude=['int', 'float']).columns)
    data[categorical_columns] = data[categorical_columns].fillna('Missing')
    data.fillna(-999, inplace=True)

    return data
```

```python
preprocessed = preprocessing(data)
preprocessed['SalePrice'] = np.log(preprocessed['SalePrice'])
nan_counts = preprocessed.isnull().sum(axis = 0)
nan_counts[nan_counts > 0]
```

```python
y = preprocessed['SalePrice']
x = preprocessed.drop('SalePrice', axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_validation, y_train, y_validation = train_test_split(x, y, train_size=0.75, random_state=42)
```

```python
categorical_features_indices = list(x.select_dtypes(exclude=['int', 'float']).columns)
```

```python
from catboost import Pool

train_pool = Pool(x_train, y_train, cat_features=categorical_features_indices)
validation_pool = Pool(x_validation, y_validation, cat_features=categorical_features_indices)
pool = Pool(x, y, cat_features=categorical_features_indices)
```

```python
cat_params = {
    'loss_function': 'RMSE',
    'use_best_model': True,
    'od_type': 'Iter',
    'od_wait': 40
    #'task_type': 'GPU'
}
```

```python
from catboost import CatBoostRegressor

cat = CatBoostRegressor(**cat_params)
```

```python
cat.fit(train_pool, eval_set=validation_pool)#, save_snapshot=True)
```

```python
cat.score(train_pool)
```

```python
cat.score(validation_pool)
```

```python
feature_importances = cat.get_feature_importance(train_pool)
feature_names = x_train.columns
for score, name in sorted(zip(feature_importances, feature_names), reverse=True):
    print('{}: {}'.format(name, score))
```

```python
from catboost import cv

cv_params = cat.get_params()

cv_data = cv(
    train_pool,
    cv_params,
    logging_level='Silent'
)
```

```python
print('Best validation RMSE score: {:.2f}±{:.2f}'.format(
    np.min(abs(cv_data['test-RMSE-mean'])),
    cv_data['test-RMSE-std'][np.argmin(abs(cv_data['test-RMSE-mean']))]
))
```
