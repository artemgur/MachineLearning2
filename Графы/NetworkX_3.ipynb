## Задание 1 (1 балл)
Создайте граф и добавьте в него $5$ вершин с городами России, у каждой должен быть атрибут (характеристика) `time_diff`, означающий временную разницу с Москвой в часах.

```python
g = nx.Graph()
g.add_node('Москва', time_diff=0)
g.add_node('Казань', time_diff=0)
g.add_node('Самара', time_diff=1)
g.add_node('Красноярск', time_diff=4)
g.add_node('Владивосток', time_diff=7)
g
```

```python
nx.draw(g, with_labels = True)
```

## Задание 2 (1 балл)
Выберите какую-нибудь конкретную авиакомпанию (Аэрофлот, S7, Utair, etc).

В созданный граф добавьте рёбра между выбранными $5$ю городами, между которыми есть авиарейсы данной компании. У каждого ребра должны быть следующие атрибуты:
- расстояние между городами в километрах `distance`
- время перелёта в минутах `time`

При желании можно добавить в граф также несколько городов, являющихся пересадочными пунктами.

```python
g.add_edge('Москва', 'Казань', time=95, distance=720)
g.add_edge('Москва', 'Самара', time=115, distance=856)
g.add_edge('Москва', 'Красноярск', time=280, distance=3363)
g.add_edge('Москва', 'Владивосток', time=495, distance=6434)
g.add_edge('Красноярск', 'Владивосток', time=300, distance=3111)
```

```python
nx.draw(g, with_labels = True)
```

## Задание 3 (1 балл)
Выберите любую пару городов и постройте между ними следующие кратчайшие пути:
- без взвешивания
- со взвешиванием по расстоянию
- со взвешиванием по времени

Поясните полученные результаты.

Кратчайший путь без взвешивания:

```python
nx.shortest_path(g, 'Владивосток', 'Москва')
```

Кратчайший путь со взвешиванием по расстоянию:

```python
nx.shortest_path(g, 'Владивосток', 'Москва', weight='distance')
```

Кратчайший путь со взвешиванием по времени:

```python
nx.shortest_path(g, 'Владивосток', 'Москва', weight='time')
```

Во всех случаях прямой рейс оказался короче рейса с пересадкой.
